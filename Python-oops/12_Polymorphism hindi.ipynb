```python
class data_science:

    def syllabus(self) :
        print("this is my syllabus for data science masters ")
```

```python
class web_dev:

    def syllabus(self) :
        print("this is  my subllabus for web dev" )
```

```python
def class_parcer(class_obj):
    for i in class_obj :
        i.syllabus()
```

```python
data_science = data_science()
```

```python
web_dev = web_dev()
```

```python
class_obj = [data_science , web_dev]
```

```python
class_parcer(class_obj)
```

OOPS


```python
# Encapsulation
class BankAccount:
    def __init__(self):
        self.__balance = 10000  # Private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount()
acc.deposit(5000)
print(acc.get_balance())
```

Abstraction

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car Started")

c = Car()
c.start()
```

Inheritance

```python
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    def bark(self):
        print("Barking")

d = Dog()
d.eat()   # Parent class method
d.bark()  # Child class method
```

Polymorphism

```python
class UPI:
    def pay(self):
        print("Payment through UPI")

class Card:
    def pay(self):
        print("Payment through Card")

for payment in [UPI(), Card()]:
    payment.pay()
```
