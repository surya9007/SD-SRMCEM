```python
class car :
    def __init__(self , year , make , model, speed ) :

        self.__year = year
        self.__make = make
        self.__model = model
        self.__speed = 0

    def set_speed(self,speed) :
        self.__speed = 0 if speed < 0 else speed

    def get_speed(self)  :
        return self.__speed
```

```python
c = car(2021 , "toyata" , "innova" , 12)
```

```python
c._car__year
```

```python
c.set_speed(-3245)
```

```python
c.get_speed()
```

```python
class bank_account :
    def __init__(self , balance) :
        self.__balanace = balance

    def deposit(self , amount) :
        self.__balanace = self.__balanace + amount

    def withdaw(self ,amount) :
        if self.__balanace >= amount :
            self.__balanace = self.__balanace - amount
            return True
        else :
            return False

    def get_banalnce(self):
        return self.__balanace
```

```python
sudh = bank_account(1000)
```

```python
sudh.get_banalnce()
```

```python
sudh.deposit(5000)
```

```python
sudh.get_banalnce()
```

```python
sudh.deposit(1000)
```

```python
sudh.get_banalnce()
```

```python
sudh.withdaw(9000)
```

```python
sudh.withdaw(2000)
```

```python
sudh.get_banalnce()
```
